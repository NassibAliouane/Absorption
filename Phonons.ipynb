```python
x = np.linspace(-2,2)
y = -25*x**2 
y2 = 25*x**2 + 25
```

```python
plt.plot(x,y,label = 'Bande de valence')
plt.plot(x,y2,label = 'Bande de conduction directe')
plt.plot(x+2,y2,label = 'Bande de conduction indirecte')
#plt.axhline(0)
plt.plot((0,0),(0,25))
plt.plot((0,2),(0,25))
plt.arrow(x=0, y=0, dx=2, dy=0, width=.08) 
plt.xticks([])
plt.yticks([])
plt.xlabel('\np')
plt.ylabel('E         ',rotation = 'horizontal')
plt.legend(loc='lower right')
plt.text(1,-10,'$\Delta p$')
plt.text(-0.6,15,'$E_{directe}$')
plt.text(0.5,20,'$E_{indirecte}$')
plt.savefig('Phonon')
```
